### Fetching data from MySQL database using pandas, mysql connector and sqlalchemy

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine("MYSQL_ENGINE_URL")
```

```python
conn = engine.connect()
```

```python
query = "SELECT * FROM visa_dataset"
```

```python
data = pd.read_sql(query, engine)
data.head(10)
```

```python
conn.close()
```
